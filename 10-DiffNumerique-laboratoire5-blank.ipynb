### Exercice 1.4 — Tableau de convergence

Pour vérifier l'ordre de convergence $\mathcal{O}(h)$ d'Euler, on calcule l'**erreur maximale** sur tout l'intervalle $[0, T]$ pour plusieurs valeurs de $h$, puis le **rapport** des erreurs consécutives.

Si la méthode est d'ordre 1, diviser $h$ par 2 doit diviser l'erreur par environ 2. Une façon de voir cela est de calculer:

$$\text{rapport} = \frac{E(2h)}{E(h)} \approx 2.$$

> **Nouvelle notion Python — liste et `np.abs` :**
> ```python
> valeurs_h = [0.5, 0.25, 0.125, 0.0625, 0.03125]  # liste de pas

>     erreur = np.max(np.abs(y_approx - y_exacte))  # erreur max sur [0,T]

> `np.abs` calcule la valeur absolue élément par élément, `np.max` retourne le maximum du tableau.

## Partie 3 — Méthode de Runge-Kutta d'ordre 4 (RK4)

### Rappel de l'algorithme

À chaque pas $i$, on calcule quatre pentes :

$$k_1 = f(t_i,\; y_i)$$
$$k_2 = f\!\left(t_i + \tfrac{h}{2},\; y_i + \tfrac{h}{2}k_1\right)$$
$$k_3 = f\!\left(t_i + \tfrac{h}{2},\; y_i + \tfrac{h}{2}k_2\right)$$
$$k_4 = f\!(t_i + h,\; y_i + h\,k_3)$$

puis :

$$y_{i+1} = y_i + \frac{h}{6}(k_1 + 2k_2 + 2k_3 + k_4), \qquad t_{i+1} = t_i + h.$$
