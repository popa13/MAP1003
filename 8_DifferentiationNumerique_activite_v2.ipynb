### Rappel des formules

Soit $f$ une fonction et $h > 0$ un paramètre de pas.

**Dérivée première :**

| Nom | ------------Formule -----------------------| Ordre |
|---|---|---|
| Différence avant | $\displaystyle T_h = \frac{f(x_0+h)-f(x_0)}{h}$ | 1 |
| Différence arrière | $\displaystyle T_h = \frac{f(x_0)-f(x_0-h)}{h}$ | 1 |
| Différence avant | $\displaystyle T_h = \frac{-f(x_0+2h)+4f(x_0+h)-3f(x_0)}{2h}$ | 2 |
| Différence centrée | $\displaystyle T_h = \frac{f(x_0+h)-f(x_0-h)}{2h}$ | 2 |
| Différence arrière | $\displaystyle T_h = \frac{3f(x_0)-4f(x_0-h)+f(x_0-2h)}{2h}$ | 2 |

**Dérivée seconde (différence centrée, ordre 2) :**
$$T_h = \frac{f(x_0+h) - 2f(x_0) + f(x_0-h)}{h^2}$$

**Extrapolation de Richardson** (pour une méthode $T_h$ d'ordre $p$) :
$$R_h = \frac{2^p\,T_{h/2} - T_h}{2^p - 1}$$

**Estimation numérique de l'ordre :**
$$p \approx -\frac{\log\!\left(\dfrac{e_{h/2}}{e_h}\right)}{\log 2}, \quad\text{où }\; e_h = |f'(x_0) - T_h|$$

## Partie 2 — Tableau de convergence et ordre d'une méthode

Si une méthode $T_h$ est d'ordre $p$, alors en divisant $h$ par 2, l'erreur $e_h = |f'(x_0) - T_h|$ est divisée par $2^p$.  
On peut donc **estimer l'ordre** à partir du tableau de convergence grâce à la formule :

$$p \approx -\frac{\log\!\left(\dfrac{e_{h/2}}{e_h}\right)}{\log 2}$$

On travaille avec $f(x) = \cos(x)$, $x_0 = \pi/4$.  
La valeur exacte est $f'(\pi/4) = -\sin(\pi/4) = -\sqrt{2}/2 \approx -0.7071$.

### 🛠️ Boîte à outils Python — Boucles `for` et listes

Une boucle `for` répète une opération pour chaque élément d'une liste :

```python

    print(f"h = {h}")
# Affiche : h = 1, puis h = 0.5, puis h = 0.25, puis h = 0.125
```

Pour construire le tableau, on a besoin de mémoriser l'erreur du tour précédent.  
Voici le patron à utiliser :


## Partie 4 — Extrapolation de Richardson

L'extrapolation de Richardson permet d'obtenir une **meilleure approximation** en combinant deux estimations déjà calculées, sans avoir besoin d'un pas $h$ plus petit.

Pour une méthode $T_h$ d'ordre $p$, on calcule :
$$R_h = \frac{2^p\,T_{h/2} - T_h}{2^p - 1}$$

Le résultat $R_h$ est **au moins d'ordre $p+1$** (théorème du cours).

### 🛠️ Boîte à outils Python — Opérations sur les nombres

Pour calculer $2^p$ en Python :
```python
p = 1
print(2**p)     # affiche 2

p = 2
print(2**p)     # affiche 4
```

La formule $R_h = \dfrac{2^p\,T_{h/2} - T_h}{2^p - 1}$ s'écrit donc :
```python
def richardson(Th, Th_demi, p):
    return (2**p * Th_demi - Th) / (2**p - 1)
```
où `Th` est $T_h$, `Th_demi` est $T_{h/2}$ et `p` est l'ordre de la méthode.

Pour calculer le logarithme en base 2 :
```python
np.log2(0.25)   # = -2.0  (car 2^(-2) = 0.25)

# Ou avec le logarithme naturel :
np.log(0.25) / np.log(2)   # même résultat
```
