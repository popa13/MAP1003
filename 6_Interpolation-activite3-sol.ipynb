# Table des différences divisées

D'après la définition, la énième différence divisée d'une fonction est

Screenshot from 2026-02-22 11-37-58.png

Nous allons maintenant construire le tableau des différences divisées.


```python
def differences_divisees(x,y):
  # On enregistre le nombre de points d'interpollation
  n = len(y)

  # On initialise la table des differences divisees
  diff_div = np.zeros([n,n])

  # On initialise la premiere colonne du tableau comme etant y
  diff_div[:,0] = y

  # On construit maintenant le tableau
  for j in range(1, n):
    for i in range(n - j):
      diff_div[i,j] = (diff_div[i+1,j-1] - diff_div[i,j-1]) / (x[i+j] - x[i])

  return diff_div
```

# Schéma de Horner
Avant de définir la fonction pour déterminer le polynôme d'interpolation de Newton passant par les n + 1 points d'interpolation, on définit la fonction qui permet d'évaluer un polynôme de manière efficace.

Veuillez compléter le code suivant de la fonction *schema_horner_newton*. Après avoir ajouter le code demandé, assurez-vous d'effacer les commande **pass**.

```python
def schema_horner_newton(coef, x_data, x_eval):
  n = len(coef) - 1
  # Initialiser p
  p = coef[n]
  for i in range(n-1, -1, -1):
    # appliquer la formule de la ligne 5 de l'algorithme 6.2 en remplacant x and x - x_i
    p = coef[i] + (x_eval - x_data[i]) * p
  return p
```

# Système tridiagonaux
Nous allons maintenant implémenter la [méthode de Thomas](https://en.wikipedia.org/wiki/Tridiagonal_matrix_algorithm) pour résoudre un système tridiagonale de la forme

A x = b .

Cet algorithme permet d'utilise moins d'espace mémoire en représentant la matrice tridiagonale A du système d'équations en ne préservant que les éléments non nuls de la matrice A. Cet algorithme résout le système en O(n) opérations.

La matrice tridiagonale du système est A qu'on transforme en trois vecteurs a, d, c pour sauver de l'espace mémoire. Le vecteur a est la sous-diagonale, le vecteur d est la diagonale et le vecteur c est la sur-diagonale.

```python
def solve_tridiagonale(a,d,c,b):
  n = len(b)
  x = np.zeros(n)
  w = 0

  # Phase d'elimination
  for i in range(1,n-1):
    w = a[i-1] / d[i-1]
    d[i] = d[i] - w * c[i-1]
    b[i] = b[i] - w * b[i-1]

  w = a[n-2] / d[n-2]
  d[n-1] = d[n-1] - w * c[n-2]
  b[n-1] = b[n-1] - w * b[n-2]

  # Phase de substitution à rebours
  x[n-1] = b[n-1] / d[n-1]
  for i in range(n-2, -1, -1):
    x[i] = (b[i] - c[i] * x[i + 1]) / d[i]

  return x
```

```python
from scipy.interpolate import CubicSpline

# Les points d'interpolation
t = np.array([0, 5, 10, 15, 20, 25, 30, 35, 40, 45])
v = np.array([55, 60, 58, 54, 55, 60, 54, 57, 52, 49])

# La spline cubique
cs = CubicSpline(t,v,bc_type="natural")

# Graphe de la spline cubique
xmin, xmax = np.min(t), np.max(t)
x_values = np.linspace(xmin-1, xmax+1, 200)
y_interp = cs(x_values)
plt.plot(x_values, y_interp, color="blue", label="Spline cubique")
plt.scatter(t, v, color="red", label="Points d'interpolation")
plt.ylim(20, 70)
plt.xticks(np.arange(0, 51, 5))
plt.yticks(np.arange(20, 71, 5))
plt.grid(True)
plt.legend()
plt.show()

```
